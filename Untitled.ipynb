```python
import torch
from torch import optim
from tqdm import tqdm
# Imports for this repository
from gmm_data_generator import load_db
from numpy_dataset import NumpyDataset
import torch.nn as nn
import torchvision.transforms as transforms
import numpy as np
from random import choices
from matplotlib import pyplot as plt
```

```python
PARAMS = {'batch_size': 512,
          'zdim': 2,
          'gamma':0.5,
          'beta1':0.5,
          'beta2':0.999,
          'epochs':20,
          'lr_d': 15e-5,
          'lr_g': 15e-4,
          'epsilon': 1e-8,  # for avoiding numerical instabilities
          'samp_num_gen': 2500}
```

```python
training_data = load_db(512, 0.01)
train_dataset = NumpyDataset(training_data)
train_loader = torch.utils.data.DataLoader(dataset=train_dataset,
                                           batch_size=PARAMS['batch_size'],
                                           shuffle=True)
```

```python
from matplotlib import pyplot as plt

plt.plot(training_data[:, 0], training_data[:, 1], 'o',label='real data')
plt.show()
```

```python
import seaborn as sns
samps = training_data
cols = 1
bg_color  = sns.color_palette('Greens', n_colors=256)[0]
plt.figure(figsize=(2*10, 4*(cols/5)+5))
for i in range(1):
    if i==0:
        ax = plt.subplot((cols/5)+1,5,1)  
        plt.ylabel('DO-GAN')  
    else:
        plt.subplot((cols/5)+1,5,i+1, sharex=ax, sharey=ax)
    ax2 = sns.kdeplot(samps[:, 0], samps[:, 1], shade=True, cmap='Greens', n_levels=20, clip=[[-3,3]]*2)
    ax2.set_facecolor(bg_color)
    plt.xticks([]); plt.yticks([])
    plt.title('iteration %d'%(i*1000))
plt.gcf().tight_layout()
```

```python
glist = []
glist.append(training_data)
np_samples_ = glist
cols = len(np_samples_)
bg_color  = sns.color_palette('Greens', n_colors=256)[0]
plt.figure(figsize=(2*5, 2*(cols/5)+1))
for i, samps in enumerate(np_samples_):
    if i==0:
        ax = plt.subplot((cols/5)+1,5,1)  
        plt.ylabel('DO-GAN')  
    else:
        plt.subplot((cols/5)+1,5,i+1, sharex=ax, sharey=ax)
    ax2 = sns.kdeplot(samps[:, 0], samps[:, 1], shade=True, cmap='Greens', n_levels=20, clip=[[-3,3]]*2)
    ax2.set_facecolor(bg_color)
    plt.xticks([]); plt.yticks([])
    plt.title('iteration %d'%(i*1000))
plt.gcf().tight_layout()
```
